# Libraries

```python
from Bio.Seq import Seq
import matplotlib.pyplot as plt
```

# Data

```python
SeqA = 'AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC'
SeqB = 'CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG'
```

# Quest
Please use Biopython to:<br>
•	Find the length of both sequences.<br>
•	Determine the number of times the codon (CGC) appears in both sequences.<br>
•	Locate the first occurrence index position of the sub-sequence (CAGTC) in both sequences.<br>
•	Combine the first 15 nucleotides from sequence A and the last 10 nucleotides from sequence B to create a new DNA sequence called sequence C.<br>
•	Reverse the sequence of sequence C.<br>
•	Plot the nucleotide base (A, C, G, T) frequency of each of the sequence A, sequence B, and sequence C using the Matplotlib library.


### Length of both sequence

```python
print(f'Length of Sequence A: {len(SeqA)}')
print(f'Length of Sequence B: {len(SeqB)}')
```

### Determine the number of times the codon (CGC) appears in both sequence.

```python
codon = 'CGC'
print(f'Number of CGC appear in SeqA: {SeqA.count(codon)}')
print(f'Number of CGC appear in SeqB: {SeqB.count(codon)}')
```

### Locate the first occurance index position of sub-sequence (CAGTC) in both sequence.

```python
subseq = 'CAGTC'
print(f'First occurance index position CAGTC in SeqA: {SeqA.find(subseq)}')
print(f'First occurance index position CAGTC in SeqB: {SeqB.find(subseq)}')
```

### Combine the first 15 nucleotides from sequence A and the last 10 nucleotides from sequence B to create a new DNA sequence called sequence C.

```python
SeqC = SeqA[:15] + SeqB[-10:]
print(SeqC)
```

### Reverse the Sequence of Sequence C

```python
reversed_c = SeqC[::-1]
print(reversed_c)
```

### Plot the nucleotide base (A, C, G, T) frequency of each of the sequence A, sequence B, and sequence C using the Matplotlib library.


```python
from collections import Counter

CounterA = Counter(SeqA)
CounterB = Counter(SeqB)
CounterC = Counter(SeqC)
```

#### Sequence A

```python
keyA = CounterA.keys()
valuesA = CounterA.values()
plt.bar(keyA, valuesA)
plt.title('Nucleotide base for Sequence A')
plt.show()
```

#### Sequence B

```python
keyB = CounterB.keys()
valuesB = CounterB.values()
plt.title('Nucleotide base for Sequence B')
plt.bar(keyB, valuesB)
plt.show()
```

#### Sequence C

```python
keyC = CounterC.keys()
valuesC = CounterC.values()
plt.title('Nucleotide base for Sequence C')
plt.bar(keyC, valuesC)
plt.show()
```
