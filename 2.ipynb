# Libraries

```python
from Bio import SeqIO
import matplotlib.pyplot as plt
from Bio.SeqUtils import MeltingTemp as MT, molecular_weight, gc_fraction as GC
```

# Data

```python
SeqA = 'AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC'
SeqB = 'CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG'
```

# Quest
Please use Biopython to:<br>
•	Determine the molecular weight of both sequence.<br>
•	Refer to the Wallace rule to calculate the melting point of sequence A and sequence B and plot them using Matplotlib library.<br>
•	Calculate the GC content of sequence A and sequence B.<br>
•	Calculate the AT content of sequence A and sequence B.<br>
•	Plot the GC and AT content of sequence A and sequence B using Matplotlib library.<br>


### Determine Molecular Weight of Both Sequence

```python
print(f'Molecular weight of SeqA: {molecular_weight(SeqA)}')
print(f'Molecular weight of SeqB: {molecular_weight(SeqB)}')
```

### Calculate Melting Point of Sequence A and B

#### Wallace Rule

```python
print(f'Melting Temp SeqA (Wallace): {MT.Tm_Wallace(SeqA)}')
print(f'Melting Temp SeqB (Wallace): {MT.Tm_Wallace(SeqB)}')
plt.figure(figsize = (12, 8))
plt.bar(x = SeqA, height=MT.Tm_Wallace(SeqA))
plt.bar(x = SeqB, height=MT.Tm_Wallace(SeqB))
plt.title('Wallace Melting Temp')
plt.show()
```

#### GC Rule

```python
print(f'Melting Temp SeqA (GC): {MT.Tm_GC(SeqA)}')
print(f'Melting Temp SeqB (GC): {MT.Tm_GC(SeqB)}')

plt.figure(figsize = (12, 8))
plt.title('GC Melting Temp')
plt.bar(x = SeqA, height= MT.Tm_GC(SeqA))
plt.bar(x = SeqB, height= MT.Tm_GC(SeqB))
plt.show()
```

#### NN Rule

```python
print(f'Melting Temp SeqA (NN): {MT.Tm_NN(SeqA)}')
print(f'Melting Temp SeqB (NN): {MT.Tm_NN(SeqB)}')

plt.figure(figsize = (12, 8))
plt.title('NN Melting Temp')
plt.bar(x = SeqA, height = MT.Tm_NN(SeqA))
plt.bar(x = SeqB, height = MT.Tm_NN(SeqB))
plt.show()
```

### Calculate The GC Content of Sequence A and Sequence B

```python
print(f'GC content of SeqA: {GC(SeqA) * 100}')
print(f'GC content of SeqB: {GC(SeqB) * 100}')
```

### Calculate the AT Content of Sequence A and Sequence B

```python
print(f'AT content of SeqA: {(1 - GC(SeqA))*100}')
print(f'AT content of SeqB: {(1 - GC(SeqB))*100}')
```

### Plot AT and GC Content using MatplotLib Library

#### AT Content

```python
plt.figure(figsize=(12, 8))
plt.title('AT Content')
plt.bar(x = SeqA, height=((1 - GC(SeqA))*100))
plt.bar(x = SeqB, height=((1 - GC(SeqB))*100))
plt.show()
```

#### GC Content

```python
plt.figure(figsize=(12, 8))
plt.title('GC Content')
plt.bar(x = SeqA, height=GC(SeqA)*100)
plt.bar(x = SeqB, height=GC(SeqB)*100)
plt.show()
```
