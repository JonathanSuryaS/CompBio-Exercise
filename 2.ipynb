```python
from Bio.Seq import Seq
from Bio.SeqUtils import molecular_weight, gc_fraction
import matplotlib.pyplot as plt

# Define the sequences
sequence_A = Seq("ATGCGCATAGCTAGCTCGATCGCAGTCGATCGCGATCGCGTACG")
sequence_B = Seq("CGTACGTAGCTAGCGTACGTAGCGTCAGTCTAGCTAGCGTACGT")

# 1. Determine the molecular weight of both sequences
mw_A = molecular_weight(sequence_A, seq_type='DNA')
mw_B = molecular_weight(sequence_B, seq_type='DNA')

# 2. Calculate the melting point using the Wallace rule
def wallace_rule(seq):
    a_count = seq.count('A')
    t_count = seq.count('T')
    g_count = seq.count('G')
    c_count = seq.count('C')
    return 2 * (a_count + t_count) + 4 * (g_count + c_count)

tm_A = wallace_rule(sequence_A)
tm_B = wallace_rule(sequence_B)

# Plot the melting points
plt.figure()
plt.bar(['Sequence A', 'Sequence B'], [tm_A, tm_B], color=['blue', 'green'])
plt.title('Melting Points of DNA Sequences')
plt.ylabel('Melting Point (°C)')
plt.show()

# 3. Calculate the GC content of both sequences
gc_content_A = gc_fraction(sequence_A) * 100
gc_content_B = gc_fraction(sequence_B) * 100

# 4. Calculate the AT content of both sequences
def at_content(seq):
    a_count = seq.count('A')
    t_count = seq.count('T')
    return (a_count + t_count) / len(seq) * 100

at_content_A = at_content(sequence_A)
at_content_B = at_content(sequence_B)

# 5. Plot the GC and AT content
plt.figure()
bar_width = 0.35
index = range(2)
gc_contents = [gc_content_A, gc_content_B]
at_contents = [at_content_A, at_content_B]

plt.bar(index, gc_contents, bar_width, label='GC Content', color='blue')
plt.bar([i + bar_width for i in index], at_contents, bar_width, label='AT Content', color='green')

plt.xlabel('Sequences')
plt.ylabel('Content (%)')
plt.title('GC and AT Content of DNA Sequences')
plt.xticks([i + bar_width / 2 for i in index], ['Sequence A', 'Sequence B'])
plt.legend()

plt.show()

# Print the results
print(f"Molecular weight of sequence A: {mw_A}")
print(f"Molecular weight of sequence B: {mw_B}")
print(f"Melting point of sequence A: {tm_A}°C")
print(f"Melting point of sequence B: {tm_B}°C")
print(f"GC content of sequence A: {gc_content_A}%")
print(f"GC content of sequence B: {gc_content_B}%")
print(f"AT content of sequence A: {at_content_A}%")
print(f"AT content of sequence B: {at_content_B}%")
```
