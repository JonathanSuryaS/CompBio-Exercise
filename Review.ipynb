## Pairwise

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
```

### XX

```python
seqBCglobalXX = pairwise2.align.globalxx(seqB, seqC)
for a in seqBCglobalXX:
    print(format_alignment(*a))

#xx -> match +1
#mx -> custom score buat match & mismatch, misal match +1 dan mis -1
#ms -> custom score match, mismatch, gap start, gap extend
```

### Ms -> Diambil paling tinggi, yang lain kemungkinan dibawah 2.6

```python
seqBCglobalms = pairwise2.align.globalms(seqB, seqC, +1, -1, -0.5, -0.1)
for a in seqBCglobalms:
    print(format_alignment(*a))
```

### Mx

```python
seqBCglobalmx = pairwise2.align.globalmx(seqB, seqC, +1, -1)
for a in seqBCglobalmx:
    print(format_alignment(*a))
```

### Hamming

```python
def hamming(seq1, seq2):
    h = []

    for x,y in zip(seq1, seq2):
        if(x!=y):
            h.append((x,y))

    return len(h)

print(hamming(seqB, seqC))
print(seqB)
print(seqC)
```

### Levenshtein

```python
from Levenshtein import distance
print(distance(seqB, seqC))

#ATTAAAGGT 1
#TTATACCTT 2
#->Perubahan, ngitung berapa perubahan yang diperlukan untuk seq2 ke seq1
```

### DotPlus

```python
def dotplus(seq1, seq2):
    print(' |' + seq2)
    print('-' * (len(seq2) + 3))

    for x in seq1:
        print(x + '|', end='')
        for y in seq2:
            print('X' if x == y else ' ', end='')
        print('')

dotplus(seqB, seqC)
```
