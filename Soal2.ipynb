# libraries

```python
from Bio.Seq import Seq
from Bio.SeqUtils import MeltingTemp, molecular_weight, gc_fraction
import matplotlib.pyplot as plt
```

# Data

```python
SeqA = 'AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC'
SeqB = 'CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG'
```

Molecular Weight

```python
print(f'Molecular Weight SeqA: {molecular_weight(SeqA)}')
print(f'Molecular Weight SeqA: {molecular_weight(SeqB)}')
```

Refer to the Wallace rule to calculate the melting point of sequence A and sequence B and plot them using Matplotlib library.

```python
print(f'Melting Point SeqA: {MeltingTemp.Tm_Wallace(SeqA)}')
print(f'Melting Point SeqB: {MeltingTemp.Tm_Wallace(SeqB)}')

plotter = ['SeqA', 'SeqB']
plt.bar(x=SeqA, height=MeltingTemp.Tm_Wallace(SeqA), edgecolor='black' )
plt.bar(x=SeqB, height=MeltingTemp.Tm_Wallace(SeqB), edgecolor='black' )
plt.show()
```

Calculate the GC content of sequence A and sequence B.

```python
print(f'GC Content of SeqA: {gc_fraction(SeqA) * 100}')
print(f'GC Content of SeqB: {gc_fraction(SeqB) * 100}')
```

Calculate the AT content of sequence A and sequence B.

```python
print(f'GC Content of SeqA: {(1 - gc_fraction(SeqA)) * 100}')
print(f'GC Content of SeqB: {(1- gc_fraction(SeqB)) * 100}')
```

Plot the GC and AT content of sequence A and sequence B using Matplotlib library.

```python
plt.bar("SeqA", gc_fraction(SeqA) * 100, edgecolor='black' )
plt.bar('SeqB', gc_fraction(SeqB) * 100, edgecolor='black' )
plt.show()
```

```python
plt.bar('SeqA', 100-gc_fraction(SeqA)*100, edgecolor='black')
plt.bar('SeqB', 100-gc_fraction(SeqB)*100, edgecolor='black' )
plt.show()
```
