# SubSoal 1

## Data

```python
SeqA = Seq('AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC')
SeqB = Seq('CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG')
```

## Quiz

### 1. Find the length of both sequences.

```python
print(f"The length of Sequence A: {len(SeqA)}")
print(f"The length of Sequence B: {len(SeqB)}")
```

### 2. Determine the number of times the codon (CGC) appears in both sequences.

```python
print(f"Number of times codon CGC appear in SeqA: {SeqA.count('CGC')}")
print(f"Number of times codon CGC appear in SeqB: {SeqB.count('CGC')}")
```

### 3. Locate the first occurrence index position of the sub-sequence (CAGTC) in both sequences.

```python
print(f"First occurance index position of subseq 'CAGTC' in SeqA: {SeqA.find('CAGTC')}")
print(f"First occurance index position of subseq 'CAGTC' in SeqB: {SeqB.find('CAGTC')}")
```

### 4. Combine the first 15 nucleotides from sequence A and the last 10 nucleotides from sequence B to create a new DNA sequence called sequence C.

```python
SeqC = SeqA[:15] + SeqB[-10:]
print(f"First 15 nucleotides from sequence A: {SeqA[:15]}")
print(f"Last 10 nucleotides from sequence B: {SeqB[-10:]}")
print(f"Sequence C: {SeqC}")
```

### 5. Reverse Sequence C

```python
revC = SeqC[::-1]
print(f'Sequence C: {SeqC}')
print(f"Reversed C: {revC}")
```

### 6. Plot the nucleotide base (A, C, G, T) frequency of each of the sequence A, sequence B, and sequence C using the Matplotlib library.

```python
from collections import Counter
ca = Counter(SeqA)
cb = Counter(SeqB)
cc = Counter(SeqC)
```

#### Sequence A

```python
plt.bar(ca.keys(), ca.values())
plt.title('Sequence A nucleotide')
plt.show()
```

#### Sequence B

```python
plt.bar(cb.keys(), cb.values())
plt.title('Sequence B nucleotide')
plt.show()
```

#### Sequence C

```python
plt.bar(cc.keys(), cc.values())
plt.title('Sequence C nucleotide')
plt.show()
```

```python
from Bio.pairwise2 import format_alignment
from Bio import pairwise2
```

## Data

```python
SeqA = Seq('AAGCTTAGCTAGAGCAGTCAGCTAGCTAGCT')
SeqB = Seq('TCTGAAGCTAGCTAGTGCATAGCTAGCTAGC')
```

## Quiz

### 1. Determine score of best local alignment.

```python
local_alignment = pairwise2.align.localxx(SeqA,SeqB, score_only = True)
print(local_alignment)
local_alignment = pairwise2.align.localxx(SeqA,SeqB)
for l in local_alignment:
    print(format_alignment(*l))
```

### 2. Hamming distance

```python
def hamming(SeqA, SeqB):
    h = 0
    for x,y in zip(SeqA, SeqB):
        if x!=y:
            h+=1
    return h
print(f"Hamming distance of SeqA and SeqB: {hamming(SeqA, SeqB)}")
```

### 3. Levensthein

```python
from Levenshtein import distance
print(f"Levenshtein distance of SeqA and SeqB: {distance(SeqA, SeqB)}")
```

### 4. Dotplus

```python
def dot_plus(SeqA, SeqB):
    print(' |' + SeqB)
    print('-' * (len(SeqB) + 2))
    for x in SeqA:
        print(x + '|', end = '')
        for y in SeqB:
            if x == y:
                print('X', end = '')
            else:
                print(' ', end = '')
        print('')
            
dot_plus(SeqA,SeqB)
```

```python
def pair(a, b):
    return True if a == b else False
```

```python
def create_matrix(SeqA, SeqB):
    matrix = []
    for x in SeqA:
        row = []
        for y in SeqB:
            row.append(pair(x,y))
        matrix.append(row)
    return matrix

plt.imshow(np.array(create_matrix(SeqA, SeqB)))
plt.xticks(np.arange(len(SeqB)), SeqB)
plt.yticks(np.arange(len(SeqA)), SeqA)
plt.show()
```
