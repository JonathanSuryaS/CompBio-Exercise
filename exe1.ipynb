# Libraries

```python
from Bio.Seq import Seq
import matplotlib.pyplot as plt
from collections import Counter
```

# Data

```python
SeqA = Seq('AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC')
SeqB = Seq('CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG')
```

# Question

### FInd the length of both sequence

```python
print(f"SeqA Length: {len(SeqA)}")
print(f"SeqB Length: {len(SeqB)}")
```

### Determine the number of codon CGC appearn in both sequence

```python
print(f"Number of CODON CGC appear in SeqA: {SeqA.count('CGC')}")
print(f"Number of CODON CGC appear in SeqB: {SeqB.count('CGC')}")
```

### Locate the First Occurance Index position of sub - sequence CAGTC in both sequence

```python
print(f"First Occurance index CAGTC at SeqA: {SeqA.find('CAGTC')}")
print(f"First Occurance index CAGTC at SeqB: {SeqB.find('CAGTC')}")
```

### Combine the first 15 nucleotides from sequence A and the last 10 nucleotides from sequence B to create a new DNA sequence called sequence C.

```python
SeqC = SeqA[:15] + SeqB[-10:]
print(SeqC)
```

### Reverse the sequence of sequence C

```python
reversedC = SeqC[::-1]
print(reversedC)
```

### Plot the nucleotide base (A, C, G, T) frequency of each of the sequence A, sequence B, and sequence C using the Matplotlib library.

```python
counterA = Counter(SeqA)
counterB = Counter(SeqB)
counterC = Counter(SeqC)
```

#### SeqA

```python
keysA = counterA.keys()
valuesA = counterA.values()

plt.bar(keysA, valuesA, color = ['red', 'green', 'blue', 'yellow'])
plt.title('SeqA Nucleotide')
plt.show()
```

#### SeqB

```python
keysB = counterB.keys()
valuesB = counterB.values()

plt.bar(keysB, valuesB, color = ['red', 'green', 'blue', 'yellow'])
plt.title('SeqB Nucleotide')
plt.show()
```

#### SeqC

```python
keysC = counterC.keys()
valuesC = counterC.values()

plt.bar(keysC, valuesC, color = ['red', 'green', 'blue', 'yellow'])
plt.title('SeqC Nucleotide')
plt.show()
```
