### Refer to the Wallace rule to calculate the melting point of sequence A and sequence B and plot them using Matplotlib library.

```python
import matplotlib.pyplot as plt
from Bio.SeqUtils import gc_fraction

def calculate_gc_at_content(sequence):
    gc_content = gc_fraction(sequence) * 100
    at_content = 100 - gc_content
    return gc_content, at_content

# Define the sequences
SeqA = 'AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC'
SeqB = 'CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG'

# Calculate GC and AT content for each sequence
gc_content_A, at_content_A = calculate_gc_at_content(SeqA)
gc_content_B, at_content_B = calculate_gc_at_content(SeqB)

# Plotting the GC and AT content for both sequences
labels = ['GC Content', 'AT Content']
contents_A = [gc_content_A, at_content_A]
contents_B = [gc_content_B, at_content_B]

# Create a bar chart
x = range(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(x, contents_A, width, label='SeqA', color='orange')
ax.bar([p + width for p in x], contents_B, width, label='SeqB', color='blue')

# Add labels and title
ax.set_xlabel('Content Type')
ax.set_ylabel('Percentage')
ax.set_title('GC and AT Content in DNA Sequences')
ax.set_xticks([p + width / 2 for p in x])
ax.set_xticklabels(labels)
ax.set_ylim(0, 100)
ax.legend()

plt.show()
```
