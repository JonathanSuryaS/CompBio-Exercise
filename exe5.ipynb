# Libraries

```python
from Bio.Seq import Seq
```

# Data 

```python
SeqA = Seq('AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC')
SeqB = Seq('CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG')
```

# Quiz

## 1. Find the length of both sequences.

```python
print(f"The length of SeqA: {len(SeqA)}")
print(f"The length of SeqB: {len(SeqB)}")
```

## 2. Determine the number of times the codon (CGC) appears in both sequences.

```python
print(f"Number of codon CGC appear in SeqA: {SeqA.count('CGC')}")
print(f"Number of codon CGC appear in SeqB: {SeqB.count('CGC')}")
```

## 3. Locate the first occurrence index position of the sub-sequence (CAGTC) in both sequences.

```python
print(f"First occurance index of sub - sequence CAGTC in SeqA: {SeqA.find('CAGTC')}")
print(f"First occurance index of sub - sequence CAGTC in SeqB: {SeqB.find('CAGTC')}")
```

```python
from Bio.SeqUtils import nt_search
to_find = 'CG'
#nt_seatch -> CG tu ada di index mana aja
print(f'Nt Search {to_find} in SeqA: {nt_search(str(SeqA), to_find)}')
print(f'Nt Search {to_find} in SeqB: {nt_search(str(SeqB), to_find)}')
```

## 4. Combine the first 15 nucleotides from sequence A and the last 10 nucleotides from sequence B to create a new DNA sequence called sequence C.

```python
SeqC = SeqA[:15] + SeqB[-10:]
print(f"SeqA: {SeqA}")
print(f"SeqB: {SeqB}")
print('')
print(f"First 15 nucleotide SeqA: {SeqA[:15]}")
print(f"Last 10 nucleotide SeqB: {SeqB[-10:]}")
print(f"SeqC: {SeqC}")
```

## 5. Reverse the sequence of sequence C

```python
print(f"Reversed Sequence C: {SeqC[::-1]}")
```

## 6. Plot the nucleotide base (A, C, G, T) frequency of each of the sequence A, sequence B, and sequence C using the Matplotlib library.

```python
from collections import Counter
import matplotlib.pyplot as plt
CounterA = Counter(SeqA)
CounterB = Counter(SeqB)
CounterC = Counter(SeqC)
```

### Sequence A

```python
keysA = CounterA.keys()
valuesA = CounterA.values()
plt.bar(keysA, valuesA, color = ['red', 'green', 'blue', 'yellow'])
plt.title('Sequence A Nucleotide')
plt.show()
```

### Sequence B

```python
keysB = CounterB.keys()
valuesB = CounterB.values()
plt.bar(keysB, valuesB, color = ['red', 'green', 'blue', 'yellow'])
plt.title('Sequence B Nucleotide')
plt.show()
```

### Sequence C

```python
keysC = CounterC.keys()
valuesC = CounterC.values()
plt.bar(keysC, valuesC, color = ['red', 'green', 'blue', 'yellow'])
plt.title('Sequence C Nucleotide')
plt.show()
```
