# Libraries

```python
from Bio.Seq import Seq
from Bio.SeqUtils import nt_search, molecular_weight as MW, gc_fraction as GC, MeltingTemp as MT
import matplotlib.pyplot as plt
```

# Data

```python
SeqA = Seq('AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC')
SeqB = Seq('CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG')
```

# Quiz

## 1. Determine the molecular weight of Both Sequence

```python
mwA = MW(SeqA)
mwB = MW(SeqB)

print(f"Molecular weight of Sequence A: {mwA}")
print(f"Molecular weight of Sequnece B: {mwB}")
```

## 2. Calculate the Melting Temp of Sequence A and B, and plot them using Matplotlib library

### Template

```python
# Wallace Rule
wallaceA = MT.Tm_Wallace(SeqA)
wallaceB = MT.Tm_Wallace(SeqB)

# GC Rule
gcA = MT.Tm_GC(SeqA)
gcB = MT.Tm_GC(SeqB)

# NN Rule
nnA = MT.Tm_NN(SeqA)
nnB = MT.Tm_NN(SeqB)
```

### Wallace

```python
def plot_melting(method, a, b):
    print(f"Melting Temp {method} of SeqA: {a}")
    print(f"Melting Temp {method} of SeqB: {b}")
    plt.bar('Sequence A', a)
    plt.bar('Sequence B', b)
    plt.title(f'Melting Temp of {method}')
    plt.show()
```

```python
plot_melting('Wallace', wallaceA, wallaceB)
```

```python
plot_melting('GC', gcA, gcB)
```

```python
plot_melting('NN', nnA, nnB)
```

## 3. Calculate the GC Content of Sequence A and Sequence B

```python
gc_contentA = GC(SeqA)*100
gc_contentB = GC(SeqB)*100
at_contentA = 100 - gc_contentA
at_contentB = 100 - gc_contentB
```

```python
plt.bar('Sequence A', gc_contentA)
plt.bar('Sequence B', gc_contentB)
plt.title('GC Content of Sequence A and B')
plt.show()
```

```python
print(f"GC Content Sequence A: {gc_contentA}")
print(f"GC Content Sequence B: {gc_contentB}")
```

```python
plt.bar('Sequence A', at_contentA)
plt.bar('Sequence B', at_contentB)
plt.title('AT Content of Sequence A and B')
plt.show()
```

```python
print(f"AT Content Sequence A: {at_contentA}")
print(f"AT Content Sequence B: {at_contentB}")
```
